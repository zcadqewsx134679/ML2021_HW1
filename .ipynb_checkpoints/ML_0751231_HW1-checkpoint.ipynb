```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
df = pd.read_csv(r"dataset.csv",encoding="utf-8")
```

```python
df.loc[:,'feature_2'] = df.loc[:,'feature_2'].map({'M':0,'F':1})
df.loc[:,'feature_4'] = df.loc[:,'feature_4'].map({'OD':0,'OS':1})
```

```python
df
```

```python
#處理缺失值
#df_zero  缺失值填0
#df_mean  缺失值填平均值
df_zero = df.fillna(0)
df_mean = df.copy()
for column in list(df_mean.columns[df_mean.isnull().sum() > 0]):
    mean_val = df_mean[column].mean()
    df_mean[column].fillna(mean_val, inplace=True)
```

```python
def rmse(y_true,y_pred):
    return np.sqrt(mean_squared_error(y_true,y_pred))
```

```python
# Q2(a)
X_zero = df_zero.drop(columns=['label'])
X_mean = df_mean.drop(columns=['label'])
Y_zero = df_zero.loc[:,'label']
Y_mean = df_mean.loc[:,'label']

model_zero = LinearRegression(n_jobs=-1,normalize=True)
model_zero.fit(X_zero,Y_zero)
RMSE_zero = rmse(Y_zero,model_zero.predict(X_zero))

model_mean = LinearRegression(n_jobs=-1,normalize=True)
model_mean.fit(X_mean,Y_mean)
RMSE_mean = rmse(Y_mean,model_mean.predict(X_mean))

result = pd.DataFrame(np.zeros((1,2)),columns=['fill_zero_RMSE','fill_mean_RMSE'],index = ['feature_1~118'])
result.iloc[0,0] = RMSE_zero
result.iloc[0,1] = RMSE_mean
result.to_csv('Q2_a.csv')
```

```python
#Q2(b)
#根據 Q2(a)之結果 , 空值填0之RMSE較小 故空值填0

def feature_RMSE(df,index):
    feature = df.loc[:,'feature_'+str(index)].values.reshape(-1,1)
    label = df.loc[:,'label']
    model = LinearRegression(n_jobs=-1,normalize=True)
    model.fit(feature,label)
    RMSE = rmse(label,model.predict(feature))
    return RMSE

result_RMSE = pd.DataFrame(np.zeros((119,1)),columns=['RMSE'],index = df.columns[:-1])
for i in range(0,119):
    zero_RMSE = feature_RMSE(df_zero,i)    
    result_RMSE.iloc[i,0] = zero_RMSE


result_RMSE = result_RMSE.sort_values(by=['RMSE'])
#print(result_RMSE)
result_RMSE.to_csv('Q2_b.csv')
```

```python
#Q2(c)
result_corr = df_zero.corr().loc[:,['label']]
result_corr = result_corr.sort_values(by=['label'], ascending=False)
result_corr = result_corr.drop(index = 'label')
#print(result_corr)
result_corr.to_csv('Q2_c.csv')
```

```python
#Q2(e)
feature_row = []
min_RMSE = 10
label = df.loc[:,'label']
for i in range(0,119):
    feature_row.append(result_RMSE.index[i])
    feature = df_zero[feature_row]
    model = LinearRegression(n_jobs=-1,normalize=True)
    model.fit(feature,label)
    RMSE = rmse(label,model.predict(feature))
    print("第",i,"次")
    print("RMSE: ",RMSE)
    print("min_RMSE: ",min_RMSE)
    
    if RMSE > min_RMSE :
        break
    else:
        min_RMSE = RMSE

print("final RMSE: ", min_RMSE) 


result = pd.DataFrame(np.zeros((1,1)),columns=['RMSE'],index = ['feature_0~118'])
result.iloc[0,0] = min_RMSE
result.to_csv('Q2_e.csv')
```
